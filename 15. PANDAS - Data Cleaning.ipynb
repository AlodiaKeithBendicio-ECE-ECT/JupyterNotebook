# DATA CLEANING 

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\arfil\Downloads\DATA ANALYST FOLDER\JUPYTER NOTEBOOK\Customer Call List.xlsx")
df
```

### Remove Duplicates

```python
df = df.drop_duplicates()
df
```

### Remove Columns that are not Needed

```python
df = df.drop(columns = 'Not_Useful_Column')
df
```

### Fixing, Standardizing, Normalizing

##### 1. Last_Name

```python
# df['Last_Name'] = df['Last_Name'].str.lstrip('...')
# df['Last_Name'] = df['Last_Name'].str.lstrip('/')
# df['Last_Name'] = df['Last_Name'].str.lstrip('_')
# df['Last_Name']     or

df['Last_Name'] = df['Last_Name'].str.lstrip('123./_')
df['Last_Name']
```

```python
df
```

##### 2. Phone_Number

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df['Phone_Number']
```

- change into string

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']
```

- reformat

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number']
```

- replace nan values with nothing

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df['Phone_Number']
```

```python
df
```

##### 3. Address

- split into columns

```python
df[["Street_Address", "State", "Zip_Code"]] = df['Address'].str.split(',', n = 2, expand = True)
df[["Street_Address", "State", "Zip_Code"]]
```

```python
df
```

```python
df = df.drop(columns = 'Address')
df
```

##### 4. Paying Customer

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No')
df['Paying Customer']
```

```python
df
```

##### 5. Do_Not_Contact

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', 'No')
df['Do_Not_Contact']
```

- replace the invalid data with blanks
- fill the blank data with blank space instead of "Nan, etc."

```python
df = df.replace('No/a', '')
df = df.fillna("")
df
```

### Filtering Down Rows of Data

- only have the list of numbers you will contact (i.e. Do_Not_Contact is 'No' or blank & have Phone_Number)

1. DROP Do_Not_Contact == 'Yes'

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace = True)

df 
```

2. DROP blank Phone_Number

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)

df 

# df = df.dropna(subset = 'Phone_Number', inplace = True)
```

- reset index

```python
df = df.reset_index(drop = True)

df
```
